```python
import pandas as pd
```

```python
tb_casos = pd.read_csv('caso.csv', parse_dates = ['date']) 
```

```python
print(tb_casos.info())
```

```python
tb_casos.head()
```

```python
tb_casos = tb_casos[tb_casos['place_type'] != 'state']
tb_casos = tb_casos.sort_values(['state', 'city_ibge_code', 'date'])
tb_casos.head(20)
```

```python
tb_casos['casos_ontem'] = tb_casos.groupby(['state', 'city'])['confirmed'].shift(1)
tb_casos['casos_confirmados_diarios'] = tb_casos['confirmed'] - tb_casos['casos_ontem']
tb_casos.head(20)
```

```python
tb_casos.to_csv('caso.csv')
```
